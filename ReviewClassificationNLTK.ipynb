# Create Training and Testing Data and do Data Cleaning

```python
import string
string.punctuation
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def textcleaning(message):
    Test_punctuation_removed = [char for char in message if char not in string.punctuation]
    Test_punctuation_removed_and_joined = ''.join(Test_punctuation_removed)
    Test_punctuation_removed_and_joined_clean = [word for word in Test_punctuation_removed_and_joined.split() if word.lower() not in stopwords.words('english')]
    return Test_punctuation_removed_and_joined_clean
```

```python
yelp_dataset_clean= yelp_dataset_1_5['text'].apply(textcleaning)
```

# TF(word) = Number of times the 'word' appears in a document / Total number of terms in the document

# IDF(word) = log_e(Total number of documents / Number of documents with the term 'word' in it). 

# Overall weight of the (WORD)= TF * IDF

```python
yelp_dataset_vectorizer
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
yelp_dataset_tfidf = TfidfTransformer().fit_transform(yelp_dataset_vectorizer)
```

```python
print(yelp_dataset_tfidf.shape)
```

```python
yelp_dataset_tfidf
```

```python
print(yelp_dataset_tfidf[:,:])
# Sparse matrix with all the values of IF-IDF
```

```python
X = yelp_dataset_tfidf
y = label

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

from sklearn.naive_bayes import MultinomialNB
NB_classification = MultinomialNB()
NB_classification.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classification.predict(X_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_train,y_predict_train))
```

```python
y_predict_test = NB_classification.predict(X_test)
y_predict_test
```

```python
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict_test,zero_division=1))
```
